```python
# import makerspace-related functions from our library
from makerspace import *
```

### Areas of interest (AOIs)

```python
# load the background image / font; create a new image
floorplan_path = os.path.join(base_path, 'Development/Visualization/outputs/EIS-floorplan.png')
img,draw = create_new_image(floorplan_path)

# draw 
draw_equipment(draw,laser_position,'laser_position',display=True)
draw_equipment(draw,soldering_position,'soldering_position',display=True)
draw_equipment(draw,sewing_position,'sewing_position',display=True)
draw_equipment(draw,printer_position,'printer_position',display=True)
draw_equipment(draw,printer_position,'3D printer',display=True)
draw_equipment(draw,collaboration_position,'collaboration_position',display=True)
draw_equipment(draw,tool_position,'tool_position',display=True)
pos = draw_equipment(draw,office_position,'office_position',display=True)

# display the figure 
fig = plt.figure(figsize=(10,15))
plt.imshow(np.array(img))
plt.axis('off')
plt.show()
```

### Plot makerspace users

```python
# create the background image
floorplan_path = os.path.join(base_path, 'Development/Visualization/outputs/EIS-floorplan.png')
img,draw = create_new_image(floorplan_path)
font = ImageFont.load_default()

# draw student positions
prev = None
for index, row in df.iterrows():
    joint = transform_coordinates(row['nose_x'],row['nose_y'])
    draw_ellipse(draw, joint, 5, 'blue')
    if prev != None: draw.line([prev,joint], fill='black')
    if index > 1000: break
    prev = joint

fig = plt.figure(figsize=(10,15))
plt.imshow(np.array(img))
plt.axis('off')
plt.show()
```
